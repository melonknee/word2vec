```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threshold = 1e-5

# Simulate a range of word frequencies from very rare to very common
freqs = np.logspace(-7, -2, 500)  # 500 points between 1e-7 and 1e-2
probs = []

for f in freqs:
    prob = (f / threshold)**0.5 + 1
    prob = threshold / f * prob
    prob = min(prob, 1.0)  # cap at 1
    probs.append(prob)

plt.figure(figsize=(8, 5))
plt.plot(freqs, probs)
plt.xscale("log")
plt.xlabel("Word Frequency")
plt.ylabel("Keep Probability")
plt.title("Subsampling Keep Probability vs Word Frequency")
plt.grid(True)
plt.show()
```
